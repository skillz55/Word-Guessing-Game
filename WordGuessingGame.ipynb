```python
import random

def choose_word():
    words = ["toyota", "cadillac", "honda", "volkswagen",
         "ford", "nissan", "hyundai", "porsche","maruti",
          "lamborghini", "bentley", "ferrari"]
    return random.choice(words)

def play_game():
    word = choose_word()
    guessed_letters = []
    tries = 12
    
    print("Welcome to the Word Guessing Game. Good Luck!")
    print("Guess the letters of the word.")
    
    while tries > 0:
        # Display the word with underscores for un-guessed letters
        display_word = ""
        for letter in word:
            if letter in guessed_letters:
                display_word += letter
            else:
                display_word += "_"
        print(display_word)
        
        guess = input("Enter a letter: ").lower()
        
        if guess in guessed_letters:
            print("You already guessed that letter earlier.")
            continue
        
        guessed_letters.append(guess)
        
        if guess in word:
            print("Correct guess!")
        else:
            tries -= 1
            print("Wrong guess! Tries left:", tries)
        
        if all(letter in guessed_letters for letter in word):
            print("Congrats! You guessed the word:", word)
            return
    
    print("Game over! You ran out of tries. The word was:", word)

play_game()
```
